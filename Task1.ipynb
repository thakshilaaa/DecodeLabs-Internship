```python
import pandas as pd
import numpy as np

np.random.seed(42)
n = 500

regions = ['North', 'South', 'East', 'West']
categories = ['Electronics', 'Clothing', 'Food', 'Furniture', 'Sports']
dates = pd.date_range(start='2023-01-01', periods=n, freq='D')

data = {
    'OrderID': range(1001, 1001 + n),
    'Date': np.random.choice(dates, n),
    'Region': np.random.choice(regions, n),
    'Category': np.random.choice(categories, n),
    'Product': ['Product_' + str(np.random.randint(1, 51)) for _ in range(n)],
    'UnitsSold': np.random.randint(1, 100, n),
    'UnitPrice': np.round(np.random.uniform(5.0, 500.0, n), 2),
    'Discount': np.round(np.random.uniform(0.0, 0.30, n), 2),
    'CustomerAge': np.random.randint(18, 70, n),
    'CustomerGender': np.random.choice(['Male', 'Female'], n),
}

data['TotalRevenue'] = np.round(
    data['UnitsSold'] * data['UnitPrice'] * (1 - data['Discount']), 2
)

df = pd.DataFrame(data)
```
